```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph
G = nx.Graph()

G.add_node(1, feature=[1, 2])
G.add_node(2, feature=[3, 4])
G.add_node(3, feature=[5, 6])

G.add_edge(1, 2)
G.add_edge(2, 3)
G.add_edge(3, 1)

pos = {node: G.nodes[node]['feature'] for node in G.nodes}
nx.draw_networkx_nodes(G, pos=pos, node_size=500, node_color='skyblue')
nx.draw_networkx_edges(G, pos=pos, edge_color='gray')
nx.draw_networkx_labels(G, pos=pos, font_color='black', font_weight='bold')
plt.axis('off')
plt.show()
```
